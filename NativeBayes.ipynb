```python
from sklearn.naive_bayes import MultinomialNB
import numpy as np

# train data
d1 = np.array([2,1,1,0,0,0,0,0,0])
d2 = np.array([1,1,0,1,1,0,0,0,0])
d3 = np.array([0,1,0,0,1,1,0,0,0])
d4 = np.array([0,1,0,0,0,0,1,1,1])
train_label = np.array([d1, d2, d3, d4])
label = np.array(['B', 'B', 'B', 'N'])

# test data
d5 = np.array([[2,0,0,1,0,0,0,1,0]])
d6 = np.array([[0,1,0,0,0,0,0,1,1]])

# call MultinomialNB
model = MultinomialNB()
# training
model.fit(train_label, label)

# test
print('Predicting class of d5: ', str(model.predict(d5)))
print('Probability of d6 in each class', model.predict_proba(d6))
```

```python
from sklearn.naive_bayes import BernoulliNB
d1[d1!=0] = 1
d2[d2!=0] = 1
d3[d3!=0] = 1
d4[d4!=0] = 1
train_label = np.array([d1, d2,d3,d4])
label = np.array(['B','B','B','N'])

d5[d5!=0] = 1
d6[d6!=0] = 1
model = BernoulliNB()
model.fit(train_label, label)

print('Predicting class of d5: ', str(model.predict(d5)))
print('Probability of d6 in each class', model.predict_proba(d6))
```
